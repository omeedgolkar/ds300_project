```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('precision', 2)
```

```python
filepath = '~/Documents/classes/ds3000/Project/Coffee Quality database from CQI/merged_data_cleaned.csv'
```

```python
data = pd.read_csv(filepath)
data.head()
data = data[(data.loc[:, 'Country.of.Origin'].notna()) &
           (data['Species'] == 'Arabica')]
data.tail()
```

```python
def features_and_target(df):
    features = df.iloc[:,21:33]
    target = df.loc[:, 'Country.of.Origin']
    return (features, target)
```

```python
features, target = features_and_target(data)
```

```python
features.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=3000)

knn = KNeighborsClassifier()

knn.fit(X=X_train, y=y_train)

predicted = knn.predict(X=X_test)

knn.score(X_test, y_test)
```
